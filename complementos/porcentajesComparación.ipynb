```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
algoritmos = ['GD', 'SGD', 'MBGD', 'AdaGrad', 'Adam', 'RMSProp']
precisiones = [80.00, 81.30, 76.55, 71.51, 79.02, 79.24]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(algoritmos))
plt.bar(x_pos, precisiones, color='skyblue', edgecolor='black')

# Añadir etiquetas y título
plt.xticks(x_pos, algoritmos)
plt.xlabel('Algoritmo de optimización')
plt.ylabel('Precisión (%)')
plt.title('Comparación de la precisión de diferentes optimizadores')
plt.ylim(70, 85)  # Ajustar los límites del eje Y para mejorar la visualización

# Añadir los valores encima de las barras
for i, valor in enumerate(precisiones):
    plt.text(x_pos[i], valor + 0.5, f"{valor:.2f}%", ha='center', fontsize=10)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de los algoritmos y sus precisiones
algoritmos = ["GD", "SGD", "MBGD", "AdaGrad", "Adam", "RMSProp"]
precisiones = [82.74, 82.83, 76.55, 71.51, 79.02, 79.24]

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
plt.bar(algoritmos, precisiones, color='skyblue', edgecolor='black')

# Añadir etiquetas y título
plt.xlabel("Algoritmos de Optimización", fontsize=12)
plt.ylabel("Precisión (%)", fontsize=12)
plt.title("Comparación de Precisión por Algoritmo de Optimización", fontsize=14)
plt.ylim(70, 85)  # Ajustar los límites del eje Y para mejor visualización

# Mostrar los valores exactos en la parte superior de cada barra
for i, valor in enumerate(precisiones):
    plt.text(i, valor + 0.5, f"{valor:.2f}%", ha='center', fontsize=10)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
