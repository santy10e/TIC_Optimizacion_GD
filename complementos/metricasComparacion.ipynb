```python
import matplotlib.pyplot as plt
import numpy as np

# Mtricas actualizadas
metrics = ["Precision", "Accuracy", "F1-Score", "Specificity", "Sensitivity"]
sgd_metrics = [0.803, 0.784, 0.774, 0.820, 0.747]
logistic_metrics = [0.737, 0.744, 0.744, 0.736, 0.752]

x = np.arange(len(metrics))  # Etiquetas del eje x
width = 0.35  # Ancho de las barras

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Grfica de barras para ambos modelos
bars1 = ax.bar(x - width/2, sgd_metrics, width, label="SGD", color="lightblue")
bars2 = ax.bar(x + width/2, logistic_metrics, width, label="Regresin Logstica", color="salmon")

# Configurar etiquetas y ttulo
ax.set_xlabel("Mtricas")
ax.set_ylabel("Valores")
ax.set_title("Comparacin de Mtricas entre SGD y Regresin Logstica")
ax.set_xticks(x)
ax.set_xticklabels(metrics)

# Ajustar posicin de la leyenda al fondo del grfico
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

# Agregar lneas de cuadrcula
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar los valores encima de las barras
for bars, model_metrics in zip([bars1, bars2], [sgd_metrics, logistic_metrics]):
    for bar, value in zip(bars, model_metrics):
        height = bar.get_height()
        ax.annotate(f"{value:.3f}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Desplazamiento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

# Ajustar diseo para evitar solapamientos
plt.tight_layout()
plt.show()
```
